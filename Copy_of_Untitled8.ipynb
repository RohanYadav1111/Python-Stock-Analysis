```python
import yfinance as yf
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
```

```python
tesla_data.head()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
```

```python
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
html
```

```python
tables = pd.read_html(str(soup))
tesla_revenue = tables[0]  # Extract the relevant table
```

```python
gamestop = yf.Ticker("GME")
gamestop_data = gamestop.history(period="max")
```

```python
gamestop_data.head()
```

```python
url_gme = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
html_gme = requests.get(url_gme).text

```

```python
soup_gme = BeautifulSoup(html_gme, 'html.parser')

```

```python
import matplotlib.pyplot as plt

# Plot Tesla stock price
fig, ax1 = plt.subplots()
ax1.set_xlabel('Date')
ax1.set_ylabel('Tesla Stock Price', color='tab:red')
ax1.plot(tesla_data.index, tesla_data['Close'], color='tab:red')

# Plot Tesla revenue on secondary axis
ax2 = ax1.twinx()
ax2.set_ylabel('Tesla Revenue', color='tab:blue')
ax2.plot(tesla_revenue['Date'], tesla_revenue['Revenue'], color='tab:blue')

plt.title('Tesla Stock Price and Revenue Over Time')
plt.show()
```

```python
# Plot GameStop stock price
fig, ax1 = plt.subplots()
ax1.set_xlabel('Date')
ax1.set_ylabel('GameStop Stock Price', color='tab:red')
ax1.plot(gamestop_data.index, gamestop_data['Close'], color='tab:red')

# Plot GameStop revenue on secondary axis
ax2 = ax1.twinx()
ax2.set_ylabel('GameStop Revenue', color='tab:blue')
ax2.plot(gamestop_revenue['Date'], gamestop_revenue['Revenue'], color='tab:blue')

plt.title('GameStop Stock Price and Revenue Over Time')
plt.show()
```
